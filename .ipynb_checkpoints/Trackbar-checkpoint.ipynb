```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = np.zeros((300,512,3), np.uint8)
```

```python
def testTrack(x):
    print(x)
```

```python
cv.namedWindow('image')
```

```python
cv.createTrackbar('B', 'image', 0, 255, testTrack)
cv.createTrackbar('G', 'image', 0, 255, testTrack)
cv.createTrackbar('R', 'image', 0, 255, testTrack)
```

```python
switch = '0 : OFF\n 1: ON'
cv.createTrackbar(switch, 'image', 0, 1, testTrack)
```

```python
while(1):
    cv.imshow('image', img)
    k = cv.waitKey(1) & 0xFF
    if k == 2:
        break
    
    b = cv.getTrackbarPos('B', 'image')
    g = cv.getTrackbarPos('G', 'image')
    r = cv.getTrackbarPos('R', 'image')
    s = cv.getTrackbarPos(switch,'image')
    
    if s ==0:
        img[:] = 0
    else:
        img[:] = [b,g,r]
```

```python
cv.destroyAllwindows()
```
